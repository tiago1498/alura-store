### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

def formatacao (valores):
    return f'{valores:,.2f}'.replace(',','x').replace('.',',').replace('x','.')

print(f'Faturamento\n loja 1: {formatacao(faturamento_loja)}\n loja 2: {formatacao(faturamento_loja2)}\n loja 3: {formatacao(faturamento_loja3)}\n loja 4: {formatacao(faturamento_loja4)}')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))

faturamento = [faturamento_loja, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.bar(lojas, faturamento, color='skyblue')

plt.title('Faturamento por Loja')

```

# 2. Vendas por Categoria


```python
lojas = [loja, loja2, loja3, loja4]

for i, l in enumerate(lojas, start=1):
    print(f'\nQuantidade de produtos vendidos na loja {i}:')
    print(l['Categoria do Produto'].value_counts().sort_values(ascending=False))

```

```python
import matplotlib.pyplot as plt

lojas = [loja, loja2, loja3, loja4]

for i, l in enumerate(lojas, start=1):
    categorias = l['Categoria do Produto'].value_counts()

    plt.figure(figsize=(10, 5))
    plt.bar(categorias.index, categorias.values, color='steelblue')

    plt.title(f'Produtos vendidos por categoria - Loja {i}')
    plt.xlabel('Categoria do Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

# 3. Média de Avaliação das Lojas

```python
avaliacao1 = loja['Avaliação da compra'].mean()
avaliacao2 = loja2['Avaliação da compra'].mean()
avaliacao3 = loja3['Avaliação da compra'].mean()
avaliacao4 = loja4['Avaliação da compra'].mean()

print(f' Avaliação média\n loja 1: {avaliacao1:.2f}\n loja 2: {avaliacao2:.2f}\n loja 3: {avaliacao3:.2f}\n loja 4: {avaliacao4:.2f}')
```

```python
import matplotlib.pyplot as plt

avaliacoes = [avaliacao1, avaliacao2, avaliacao3, avaliacao4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(7, 7))
plt.pie(avaliacoes, labels=lojas, autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel1.colors)
plt.title('Avaliações Médias por Loja')
plt.axis('equal')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
resultado = []
lojas = [loja, loja2, loja3, loja4]

print('Os produtos mais e menos vendidos por loja são: ')
for i, l in enumerate(lojas, start=1):
    produto_mais_vendido = l['Produto'].value_counts().idxmax()
    vendas1= l['Produto'].value_counts().max()

    produto_menos_vendido = l['Produto'].value_counts().idxmin()
    vendas2= l['Produto'].value_counts().min()

    resultado.append({
        'Loja': f'Loja {i}',
        'Produto_mais_vendido': produto_mais_vendido,
        'Vendas1': vendas1,
        'Produto_menos_vendido': produto_menos_vendido,
        'Vendas2': vendas2
 })

print(pd.DataFrame(resultado))

```

```python
df = pd.DataFrame(resultado)

plt.figure(figsize=(10, 5))
plt.plot(df['Loja'], df['Vendas1'], label='Mais vendidos', marker='o', color='blue')
plt.plot(df['Loja'], df['Vendas2'], label='Menos vendidos', marker='o', color='orange')

plt.title('Mais e Menos Vendidos por Loja')
plt.xlabel('Loja')
plt.ylabel('Quantidade')
plt.legend()
plt.grid(True)

plt.show()
```

# 5. Frete Médio por Loja


```python
frete1 = loja['Frete'].mean()
frete2 = loja2['Frete'].mean()
frete3 = loja3['Frete'].mean()
frete4 = loja4['Frete'].mean()

def formatacao (valores):
  return f'{valores:,.2f}'.replace(',', 'x').replace('.', ',').replace('x','.')

print(f" Frete médio\n loja 1: {formatacao(frete1)}\n loja 2: {formatacao(frete2)}\n loja 3: {formatacao(frete3)}\n loja 4: {formatacao(frete4)}")

```

```python
fretes = [frete1, frete2, frete3, frete4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 5))
barras = plt.bar(lojas, fretes, color=plt.cm.Pastel1.colors)


plt.title('Frete Médio por Loja')
plt.xlabel('Lojas')
plt.ylabel('Frete Médio (R$)')
plt.ylim(0, max(fretes) + 10)

for i, valor in enumerate(fretes):
    plt.text(i, valor + 0.5, formatacao(valor), ha='center', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
